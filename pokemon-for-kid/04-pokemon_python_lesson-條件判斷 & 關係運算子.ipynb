## 4️⃣ 條件判斷（Conditional Statements）


在寶可夢對戰中，如果對方是水屬性，我們就可能選擇電屬性來攻擊。
這種「根據條件來決定要做什麼」的邏輯，在程式中就叫做 **條件判斷（if statements）**。



### 🧪 if 判斷式


```python

hp = 30

if hp < 50:
    print("HP 快沒了，需要補血！")
```

### 🔁 if / else

```python

attack = 30

if attack >= 60:
    print("這張卡攻擊力很高！")
else:
    print("攻擊力普通。")
```

### 🧭 if / elif / else 多重條件

```python

type = "水"

if type == "火":
    print("使AAAA！")
elif type == "草":
    print("BBBB！")
elif type == "水":
    print("CCC！")
else:
    print("使用普通攻擊。")
```


### ⚖️ 比較運算子（Relational Operators）


| 運算子 | 說明         |
|--------|--------------|
| `==`   | 等於         |
| `!=`   | 不等於       |
| `>`    | 大於         |
| `<`    | 小於         |
| `>=`   | 大於等於     |
| `<=`   | 小於等於     |


```python
attack = 120
if attack > 100:
    print("這是傳說級攻擊！")
```
