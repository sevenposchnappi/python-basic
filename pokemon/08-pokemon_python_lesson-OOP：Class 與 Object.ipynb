## 8ï¸âƒ£ ç‰©ä»¶å°Žå‘ç¨‹å¼è¨­è¨ˆï¼šOOPï¼ˆObject-Oriented Programmingï¼‰



åœ¨ Python ä¸­ï¼Œæˆ‘å€‘å¯ä»¥ç”¨ã€Œé¡žåˆ¥ï¼ˆclassï¼‰ã€ä¾†å®šç¾©å¯¶å¯å¤¢çš„æ¨£å­ï¼Œç”¨ã€Œç‰©ä»¶ï¼ˆobjectï¼‰ã€ä¾†å‰µé€ æ¯ä¸€éš»å¯¶å¯å¤¢å¯¦é«”ã€‚

é€™å°±åƒæ˜¯ä½ å…ˆæœ‰ä¸€å¼µã€Œå¡ç‰Œæ¨¡æ¿ã€ï¼Œå†æ ¹æ“šå®ƒå°å‡ºå¤šå¼µä¸åŒçš„å¡ç‰Œï¼


```python
class PokemonCard:
    def __init__(self, name, ptype, hp, attack):
        self.name = name
        self.ptype = ptype
        self.hp = hp
        self.attack = attack

    def show(self):
        print(f"{self.name}ï¼ˆ{self.ptype}ï¼‰ - HP:{self.hp}, æ”»æ“Š:{self.attack}")

    def train(self, power):
        self.attack += power
        print(f"{self.name} çš„æ”»æ“ŠåŠ›ä¸Šå‡åˆ° {self.attack}")

pikachu_card = PokemonCard("çš®å¡ä¸˜", "é›»", 35, 55)
pikachu_card.show()
pikachu_card.train(10)
```

### ðŸ”§ å®šç¾©ä¸€å€‹å¯¶å¯å¤¢é¡žåˆ¥ï¼ˆClassï¼‰


```python

class Pokemon:
    def __init__(self, name, type, hp, attack):
        self.name = name
        self.type = type
        self.hp = hp
        self.attack = attack

    def attack_target(self, target):
        target.hp -= self.attack
        print(f"{self.name} æ”»æ“Šäº† {target.name}ï¼{target.name} çš„ HP ç¾åœ¨æ˜¯ {target.hp}")
```

### ðŸ§ª å»ºç«‹å¯¶å¯å¤¢ç‰©ä»¶ï¼ˆObjectï¼‰ä¸¦é€²è¡Œæ”»æ“Š




```python

pikachu = Pokemon("çš®å¡ä¸˜", "é›»", 100, 30)
squirtle = Pokemon("å‚‘å°¼é¾œ", "æ°´", 80, 20)

pikachu.attack_target(squirtle)
```

### ðŸ§¬ é¡žåˆ¥ç¹¼æ‰¿ï¼ˆClass Inheritanceï¼‰


```python

class FirePokemon(Pokemon):
    def flame_attack(self, target):
        damage = self.attack + 10
        target.hp -= damage
        print(f"{self.name} ä½¿ç”¨ç«ç„°æ”»æ“Š {target.name}ï¼é€ æˆ {damage} å‚·å®³ï¼Œ{target.name} å‰©ä¸‹ {target.hp} HP")

charizard = FirePokemon("å™´ç«é¾", "ç«", 120, 40)
charizard.flame_attack(pikachu)
```
