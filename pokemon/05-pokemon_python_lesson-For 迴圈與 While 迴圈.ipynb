### ðŸ”§ å®šç¾©ä¸€å€‹å¯¶å¯å¤¢é¡žåˆ¥ï¼ˆClassï¼‰

```python

class Pokemon:
    def __init__(self, name, type, hp, attack):
        self.name = name
        self.type = type
        self.hp = hp
        self.attack = attack

    def attack_target(self, target):
        target.hp -= self.attack
        print(f"{self.name} æ”»æ“Šäº† {target.name}ï¼{target.name} çš„ HP ç¾åœ¨æ˜¯ {target.hp}")
```
